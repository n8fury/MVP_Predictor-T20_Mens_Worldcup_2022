# 12. Data balancing technique(SMOT)

```python
from sklearn.preprocessing import LabelEncoder
df5 = pd.read_csv('EDA_DB.csv')
label_encoder = LabelEncoder()
label = label_encoder.fit_transform(df5.Key_Player)
df5['Target'] = label
df5['Target'].value_counts()
df5.drop(['PLAYER ID'], axis=1, inplace=True)
df5.drop(['PLAYER NAME'], axis=1, inplace=True)
df5.drop(['Team names'], axis=1, inplace=True)
df5.drop('Key_Player', axis=1, inplace=True)
```

```python
X = df5.drop(['Target'], axis=1)
y = df5['Target']

print('Feature (X) Shape Before Balancing :', X.shape)
print('Target (y) Shape Before Balancing :', y.shape)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=1)
X, y = sm.fit_resample(X, y)
print('Feature (X) Shape After Balancing :', X.shape)
print('Target (y) Shape After Balancing :', y.shape)
```
